```python
# Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importar el dataset
df = pd.read_csv('../data/Deepfake.csv')
#ver informacion del dataset
print(df.info())
```

```python
# visualizar 5 filas del dataset
df.sample(5)
```

```python
# Identificar la cantidad de valores ausentes por columna
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(df['label'])
plt.title('Distribucion de imagenes fake')
plt.ylabel('Fake o real')
plt.xlabel('Frecuencia')
plt.show()

```

```python
# vizualizar la cantidad de imagenes por genero utilizadas
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=df)
plt.title('Genero de la imagen')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', data=df)
plt.title('Grupo de edad de la imagen')
plt.xlabel('Grupo de edad')
plt.ylabel('Cantidad')
plt.show()
```
